## 차원축소

```python
df1.info()
```

```python
from sklearn.decomposition import PCA
from sklearn.manifold import TSNE
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1
```

```python
# PCA
pca = PCA(n_components=2)
result_pca = pca.fit_transform(df1.iloc[:,1:8])
df1[['X', 'Y']] = result_pca

sns.scatterplot(x = 'X', y = 'Y', data = df1)
plt.show()
```

```python
# t-SNE
tsne = TSNE(n_components=2, learning_rate = 1000, random_state = 319)
result_tsne = tsne.fit_transform(df1.iloc[:,1:8])
df1[['X', 'Y']] = result_tsne

sns.scatterplot(x = 'X', y = 'Y', data = df1)
plt.show()
```

## 실루엣계수

```python
import pandas as pd 
import numpy as np
from sklearn.cluster import KMeans
from yellowbrick.cluster import SilhouetteVisualizer
```

```python
# k를 2로 지정
kmeans_2 = KMeans(n_clusters=3, random_state=42,init='k-means++')
visualizer_2 = SilhouetteVisualizer(kmeans_2, colors='yellowbrick')
visualizer_2.fit(df1[['X', 'Y']])      
visualizer_2.show()
```

```python
from sklearn.cluster import KMeans

kmeans = KMeans(n_clusters=3, random_state=42) # 클러스터 개수 2로 설정
label = kmeans.fit_predict(df1[['X', 'Y']])
df1['Cluster'] = label

sns.scatterplot(x = 'X', y = 'Y', hue = 'Cluster', data = df1)
plt.show()
```

```python
result_df = df1.copy()

result_df.loc[:,'Cluster'] = visualizer_2.predict(df1[['X', 'Y']])
result_df.loc[:,'silhouette coefficient'] = visualizer_2.silhouette_samples_

result_df.shape
result_df.head(5)
```

```python
print('전체 데이터의 실루엣 계수 평균: %.5f' %(result_df['silhouette coefficient'].mean()))
print()
print('군집별 실루엣 계수 평균')
result_df.groupby('Cluster')['silhouette coefficient'].mean().reset_index()
```
