```python
import requests
import bs4

keyword = input()
url = f"https://www.mangoplate.com/search/{keyword}"

headers = {'User-Agent': 'Mozilla/5.0'}
raw = requests.get(url, headers=headers)

html = bs4.BeautifulSoup(raw.text,'html.parser')

target = html.find('ul',{'class':'list-restaurants'})
food = target.find_all('div',{'class':'list-restaurant-item'})

print(f"<{keyword} 맛집 베스트>")
for i in food:
    name = i.find('h2',{'class':'title'}).text
    star = i.find('strong',{'class':'point search_point'})
    ty = i.find('p',{'class':'etc'}).text
    if star:
        star = star.text
    else:
        star = i.find('strong', {'class':'point search_point expected'}).text
    print("음식점 이름 : %s평점: %s \n유형: %s" %(name, star, ty))
    print()
```
